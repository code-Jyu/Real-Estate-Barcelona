### Defining  Function to Iterate in Multiple Pages

```python
def scrap_idealista(n= 1, N = 3, url = "https://www.idealista.com/en/venta-viviendas/{one}-{two}/pagina-{n}.htm", city='barcelona',
                   province = 'barcelona', path = r"C:\Users\Jorge\Desktop\chromedriver.exe"):
    prices = []
    address = []
    description = []
    for i in range(n, N + 1):
        print("Scrapping page {}".format(i))
            
        scrapUrl = url.format(one = province, two = city, n = i)
        
        browser = webdriver.Chrome(path)
        browser.get(scrapUrl)
        sleep(10)
        soup = BeautifulSoup(browser.page_source,"lxml")
        browser.close()
        
        for container in soup.select(".item-info-container"):
            prices.append(container.select(".item-price.h2-simulated")[0].text)
            address.append(container.select('.item-link')[0].text)
            aux = ""
            for detail in container.select(".item-detail"):
                aux = aux + detail.text + " "
            description.append(aux)
    return prices, address, description
```

```python
prices, address, description = scrap_idealista(n = 1,N =30)
```

```python
df1 = pd.DataFrame({'prices':prices,'address':address,'description':description})
```

```python
df1.to_csv('../data/idealista01.csv')
```

```python
df2 = pd.DataFrame({'prices':prices,'address':address,'description':description})
```
