### Importing json File
* We will be using the json file for visializin data in tableau. 
* Our neighborhood column must much the json file information

```python
import json
```

```python
x = ""
with open("../data/map.geojson", "r", encoding='utf-8') as f:
    x = json.load(f)
```

```python
neighL = []
neigh_groupL = []

x["features"]

for neigh in x["features"]:
    neighL.append(neigh["properties"]["neighbourhood"])
    neigh_groupL.append(neigh["properties"]["neighbourhood_group"])
    
neigh_dict = dict(zip(neighL, neigh_groupL))
```

```python
def Neigh_Replace(x):
    for elem in neighL:
        if elem.lower() == x.lower():
            return elem
    return x
```

```python
df["neighborhood"] = df["neighborhood"].apply(lambda x: Neigh_Replace(x))
```
