### Concatenating Datasets
* Our final dataset will be "bcn_housing"

```python
df = pd.concat([good_condition_01,good_condition_02,new_homes,needs_ref_01,needs_ref_02])
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.reset_index(drop=True)
```

### Cleaning Data
* Extracting Information from Address and Description
* We need to create new columns with the new features

### Rooms

```python
df['rooms'] = df['description'].str.extract(r'(\d+)\sbed')
```

```python
df.head()
```

### Area

```python
df['area_m2'] = df['description'].str.extract(r'(\d+)\sm²')
```

```python
df.head()
```

### Floor

```python
df['description'] = df['description'].str.lower()
```

```python
df.head()
```

```python
df['floor'] = df['description'].str.extract(r'(\sfloor\s\d+)')
```

```python
df['ground_floor'] = df['description'].str.extract(r'(ground floor)')
```

```python
df['mezzanine'] = df['description'].str.extract(r'(mezzanine)')
```

```python
df
```

### Filling NaN in rooms column

```python
df['floor'] = df['floor'].fillna(df['ground_floor'])
```

```python
df['floor'] = df['floor'].fillna(df['mezzanine'])
```

```python
df = df.drop(['ground_floor','mezzanine'],axis=1)
```

```python
df
```

### Views

```python
df['views'] = df['description'].str.extract(r'(exterior)')
```

```python
df.head(5)
```

```python
df['views_int'] = df['description'].str.extract(r'(interior)')
```

```python
df['views'] = df['views'].fillna(df['views_int'])
```

```python
df = df.drop('views_int', axis=1)
```

```python
df
```

### Lift

```python
df['lift'] = df['description'].str.extract(r'(with\slift)')
```

```python
df['no_lift'] = df['description'].str.extract(r'(without\slift)')
```

```python
df['lift'] = df['lift'].fillna(df['no_lift'])
```

```python
df = df.drop('no_lift', axis=1)
```

```python
df.head()
```

### Flat Type

```python
df.head()
```

```python
df['type'] = df['address'].str.split().str[0]
```

```python
df
```

```python
df['city'] = df['address'].str.split().str[-1]
```

```python
df.head()
```

```python
df['neighborhood'] = df['address'].str.split('in').str[1].str.split(', Barcelona').str[0].str.split(', ').str[-1]
```

```python
df = df.drop(['address','description'], axis=1)
```

```python
df
```

### Checking Duplicates & Droping Duplicates

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
len(df)
```

### Dtypes

```python
df.dtypes
```

```python
df['prices'] = df['prices'].str.replace(r'€','')
df['prices'] = df['prices'].str.replace(r',','')
df['prices'] = pd.to_numeric(df['prices'])
```

```python
df['area_m2'] = pd.to_numeric(df['area_m2'])
```

```python
df.dtypes
```

```python
df.head(1)
```

#### We assume studios has no rooms

```python
df[(df['type'] == 'Studio') & (df['area_m2'] < 60)]['rooms'].fillna('0')
```

```python
len(df)
```

### Filtering NaN is Rooms Column

```python
df = df[df['rooms'].notna()]
```

```python
len(df)
```

### Converting Rooms Column to Numeric

```python
df['rooms'] = pd.to_numeric(df['rooms'])
```

```python
df.dtypes
```

### Checking NaN Values 

```python
df.isna().sum()
```

### Imputing NaN Values

#### Views

```python
df.loc[(df['views'].isna()) & (df['type'] != 'Flat'),'views'] = 'exterior'
```

```python
df.isna().sum()
```

#### Floor

```python
df['type'].unique()
```

```python
flat = ['Flat','Penthouse','Duplex']
```

```python
df.loc[(df['floor'].isna()) & (~df['type'].isin(flat)),'floor'] = 'Multiple'
```

```python
df.isna().sum()
```

#### Lift

```python
df.loc[df['lift'].isna() & (~df['type'].isin(flat)),'lift'] = 'without lift'
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
len(df)
```

```python
df
```

### Renaming Column Values

```python
df['lift'] = df['lift'].str.replace(r'with lift', 'Yes').replace(r'without lift', 'No')
```

```python
df = df[['city','neighborhood','condition','type','rooms','area_m2','lift','views','floor','prices']]
```

```python
df = df.reset_index(drop=True)
```

```python
df['condition'].unique()
```

```python
df['condition'] = df['condition'].str.replace(r'good_condition','good condition').replace(r'needs_refurb','needs refurb')
```

```python
df
```
